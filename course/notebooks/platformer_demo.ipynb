```python
import subprocess, sys, time, http.client, json
from pathlib import Path

GATE = Path('projects') / 'unreal_plugin' / 'http_gateway.py'
SCRIPT = Path('projects') / 'unreal_plugin' / 'examples' / 'platformer.jusu'

print('Starting minimal gateway...')
proc = subprocess.Popen([sys.executable, str(GATE), '--port', '8885'], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
time.sleep(0.5)

conn = http.client.HTTPConnection('127.0.0.1', 8885, timeout=2)
payload = json.dumps({'cmd': 'run_script_path', 'script_path': str(SCRIPT)})
conn.request('POST', '/run', body=payload, headers={'Content-Type': 'application/json'})
resp = conn.getresponse()
print('HTTP status:', resp.status)
data = json.loads(resp.read().decode('utf-8'))
print('Gateway response:')
print(json.dumps(data, indent=2))

proc.terminate()
proc.wait()
```
