```python
import importlib
import subprocess
import sys

try:
    llvmlite = importlib.import_module('llvmlite')
    print('llvmlite is available:', llvmlite.__version__)
except Exception as e:
    print('llvmlite not installed; JIT benchmarks will be skipped. Install with `pip install llvmlite` if desired.')

# If llvmlite is present, run a quick JIT benchmark (tools/jit_benchmark.py)
if 'llvmlite' in sys.modules:
    try:
        print('\nRunning jit_benchmark (this may take a few seconds)')
        r = subprocess.run([sys.executable, 'tools/jit_benchmark.py'], capture_output=True, text=True, check=True)
        print(r.stdout)
    except subprocess.CalledProcessError as e:
        print('JIT benchmark failed to run; see stderr:')
        print(e.stderr)
```
