```python
# Write a small example program to a temporary file and run it
from pathlib import Path
import subprocess, sys, tempfile

code = '''
# simple calculator: prints a+b and a*b
fn calc(a, b) {
  print("sum=", a + b)
  print("product=", a * b)
}
calc(3, 5)
'''

p = Path(tempfile.gettempdir()) / "calc_example.jusu"
p.write_text(code)
print('Wrote', p)

r = subprocess.run([sys.executable, 'compiler/jusu.py', 'run', str(p)], capture_output=True, text=True)
print('--- stdout ---')
print(r.stdout)
print('--- stderr ---')
print(r.stderr)
```
