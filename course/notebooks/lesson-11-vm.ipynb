```python
from compiler import jusu
from runtime import bytecode_compiler, vm

src = '''
fn add(a, b) { return a + b }
print(add(3, 4))
'''

print('Compiled bytecode:')
mod = jusu.parse_and_compile(src)
bc = bytecode_compiler.compile_module(mod)
print(bc)

print('\nRunning on VM:')
vm.run_bytecode(bc)
```
